````python
import pandas as pd
````

````python
companies_df = pd.read_csv("./data/largest_us_companies_2022.csv")
companies_df
````

````python
companies_df.columns
````

````python
companies_df = companies_df.rename(columns={"Rank": "rank", "Name": "name", "Name Link": "name_link",
                                            "Symbol": "symbol", "Industry": "industry",
                                            "2022 Revenue (USD millions)": "rev_usd_mil",
                                            "Revenue Growth": "rev_growth_per", "Employees": "employees",
                                            "Headquarters City": "headquarters_city", "Headquarters Link": "headquarters_link",
                                            "Headquarters Geo Loc": "headquarters_geo_loc"} )
````

````python
companies_df.columns
````

````python
companies_df[['headquarters_lat','headquarters_lon']] = companies_df['headquarters_geo_loc'].str.split(', ',expand=True)
companies_df
````

````python
companies_df = companies_df.drop('headquarters_geo_loc', axis=1)
````

````python
companies_df.columns
````

````python
companies_df['rev_usd_mil'] = companies_df['rev_usd_mil'].str.replace(',', '')
companies_df['employees'] = companies_df['employees'].str.replace(',', '')
companies_df
````

````python
stocks_df = pd.read_csv("./data/stock_prices_2022.csv")
stocks_df
````

````python
stocks_df.columns
````

````python
stocks_df = stocks_df.rename(columns={"Symbol": "symbol", "Date": "date", "Open": "open", "High": "high",
                                      "Low": "low", "Close": "close", "Adj Close": "adj_close", "Volume": "volume"} )
````

````python
stocks_df.columns
````

```sql
CREATE TABLE companies(
id INT,
rank INT,
name VARCHAR(100),
name_link VARCHAR(100),
symbol VARCHAR(10),
industry VARCHAR(100),
rev_usd_mil INT,
rev_growth_per VARCHAR(10),
employees INT,
headquarters_city VARCHAR(100),
headquarters_link VARCHAR(100),
headquarters_lat FLOAT,
headquarters_lon FLOAT
)```

```sql
CREATE TABLE stocks (
id INT,
symbol VARCHAR(10),
date DATE,
open DEC,
high DEC,
low DEC,
close DEC,
adj_close DEC,
volume INT
)```

````python
import psycopg2
from sqlalchemy import create_engine
````

````python
from config import user, password, host, port, db
````

````python
engine = create_engine(f"postgresql+psycopg2://{user}:{password}@{host}:{port}/{db}")
````

````python
companies_df.to_sql("companies",con=engine, if_exists="replace",index=False)
````

````python
from sqlalchemy import text
with engine.connect() as connection:
    result = connection.execute(text("SELECT * FROM companies"))
    for row in result:
        print(row)
````

````python
with engine.connect() as connection:
    result = connection.execute(text("SELECT COUNT(*) FROM companies"))
    for row in result:
        print(row)
````

````python
stocks_df.to_sql("stocks",con=engine, if_exists="append",index=False)
````

````python
with engine.connect() as connection:
    result = connection.execute(text("SELECT * FROM stocks LIMIT 20"))
    for row in result:
        print(row)
````

````python
with engine.connect() as connection:
    result = connection.execute(text("SELECT COUNT(*) FROM stocks"))
    for row in result:
        print(row)
````
